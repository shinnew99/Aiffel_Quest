## 2-2 Tensor

- Rank: () 
- Shape: (   )
- Type:  


```python
import tensorflow as tf
```

#### 0D Tensor(Scalar)

```python
#       (tensor), (scalar) ,   .
t0 = tf.constant(1) 
print(t0)
print(tf.rank(t0)) #   .
```

#### 1D Tensor(Vector)

```python
# 1      , (vector) ,   .
t1 = tf.constant([1, 2, 3]) 
print(t1)
print(tf.rank(t1))
```

#### 2D Tensor(Matrix)
- 2       . 
-  ,    ,  (samples) (features)   .


```python
t2 = tf.constant([[1,2,3],
                  [4,5,6],
                  [7,8,9]])
print(t2)
print(tf.rank(t2))
```

#### 3D Tensor(Vector)
- 3  (cube)     ,           . 
-  (samples), (timesteps), (features)   . 
- 3      ,       .


```python
t3 = tf.constant([[[1,2,3],
                  [4,5,6],
                  [7,8,9]],
                 [[1,2,3],
                  [4,5,6],
                  [7,8,9]],
                 [[1,2,3],
                  [4,5,6],
                  [7,8,9]]])
print(t3)
print(tf.rank(t3))
```

#### 4D Tensor
- 4  4  ,      (   3D Tensor ). 
-  (samples), (height), (width),  (channel)   .


#### 5D Tensor
- 5  5  ,    . 
-  (samples), (height), (width),  (channel)   .

## 2-3 Tensor Type  

####  

```python
# int32
i = tf.constant(2)
print(i)

# float32
f = tf.constant(2.)
print(f)

# string
s = tf.constant('Suan')
print(s)

# float16
f16 = tf.constant(2., dtype=tf.float16)
print(f16)

# int8
i8 = tf.constant(2, dtype=tf.int8)
print(i8)
```

#### 

```python
f32 = tf.cast(f16, tf.float32)
print(f32)

#  Q. 8  32   . 
i32 = tf.cast(i8, dtype=tf.int32)
print(i32)
```

#### 

```python
x = tf.constant([[1], [2], [3]])
print(x)
print(x.shape)

y = tf.reshape(x, [1, 3])
print(y)
print(y.shape)
```

#### 

```python
print(y)
print(tf.transpose(y))
print(y.shape)
```

####  

```python
print(x)
print(tf.squeeze(x))
```

#### 

```python
print(y)
print(tf.expand_dims(y, axis=0))
print(tf.expand_dims(y, axis=1))
print(tf.expand_dims(y, axis=2))
```

#### 

```python
print(x)
print(tf.split(x, 3))
```

#### 

```python
print(x)
print(tf.concat([x, x], axis=0))
print(tf.concat([x, x], axis=1))
```

## 2-4  

```python
# 0  
print(tf.constant(2) + tf.constant(2))
print(tf.constant(2) - tf.constant(2))
print(tf.add(tf.constant(2), tf.constant(2)))
print(tf.subtract(tf.constant(2), tf.constant(2)))

# , 
print(tf.constant(2) * tf.constant(2))
print(tf.constant(2) / tf.constant(2))
print(tf.multiply(tf.constant(2), tf.constant(2)))
print(tf.divide(tf.constant(2), tf.constant(2)))
```

```python
# 
print(tf.constant(2) + tf.constant(2.2))
```

```python
print(tf.cast(tf.constant(2), tf.float32) + tf.constant(2.2))
```

### 1  
#### 1  

```python
# Q. 1  2  . 
print(tf.constant([1,2,3]))
print(tf.constant([2,4,6]))
```

```python
# Q. (, , , ) . 
print( tf.constant([1,2,3]) + tf.constant([2,4,6]) )
print( tf.constant([1,2,3]) - tf.constant([2,4,6]) )
print( tf.constant([1,2,3]) * tf.constant([2,4,6]) )
print( tf.constant([1,2,3]) / tf.constant([2,4,6]) )
```

#### 2  

```python
# Q.  (2,2) 2  2  . 
a =  tf.constant([[1,2,3],
                  [4,5,6],
                  [7,8,9]])
b =  tf.constant([[9,8,7],
                  [6,5,4],
                  [3,2,1]])
print(a, b) 
```

```python
print(a + b) # element-wise addition
print(a - b) # element-wise subtraction
print(a * b) # element-wise multiplication
print(a @ b) # matrix multiplication
print(a / b) # element-wise division
```

```python
print(tf.add(a, b))
print(tf.subtract(a, b))
print(tf.multiply(a, b))
print(tf.matmul(a, b)) #  
print(tf.divide(a, b))
```

```python
c = tf.constant([[4.0, 5.0, 6.0], 
                 [10.0, 9.0, 8.0]])

print(tf.reduce_max(c))
print(tf.argmax(c))
print(tf.nn.softmax(c))
```

```python
a = tf.constant([1, 2, 3, 4, 5, 6, 7, 8]) 
b = tf.constant([[1., 2., 3., 4., 5., 6., 7., 8.]])
cast_b = tf.cast(b, tf.int32)
print(a.shape)
print(cast_b)
# Q.   2    3  .


reshaped_a = tf.reshape(a, [2, 2, 2])
reshaped_b = tf.reshape(cast_b, [2, 2, 2])

print(reshaped_a)
print(reshaped_b)


print(reshaped_a@reshaped_b)
```
