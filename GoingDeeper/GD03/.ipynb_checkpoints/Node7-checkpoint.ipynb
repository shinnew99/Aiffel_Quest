### 7-3  : -  

Step 1. 

```python
import re
from konlpy.tag import Okt
from collections import Counter
print(" ")
```

```python
text = "   !    ! ~        ."
text
```

Step 2.   <br>


```python
reg = re.compile("[^--- ]")
text = reg.sub('', text)
print(text)
```

Step 3.   <br>
     (token)    .         .

```python
okt=Okt()
tokens = okt.morphs(text)
print(tokens)
```

Step 4.   <br>
       .        .              ?   Counter      .

```python
vocab = Counter(tokens)
print(vocab)
```

```python
vocab['']
```

```python
vocab_size = 5
vocab = vocab.most_common(vocab_size) #     5  
print(vocab)
```

```python
word2idx={word[0] : index+1 for index, word in enumerate(vocab)}
print(word2idx)
```

Step 5:-   <br>
-       -  .         -   .

```python
def one_hot_encoding(word, word2index):
    one_hot_vector = [0]*(len(word2index))
    index = word2index[word]
    one_hot_vector[index-1] = 1
    return one_hot_vector
print("=3")
```

```python
one_hot_encoding("", word2idx)
```

  - (one-hot encpoding) <br>

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
print(" ")
```

```python
text = [['', '', ''],['', ''], ['', '']]
text
```

```python
t = Tokenizer()
t.fit_on_texts(text)
print(t.word_index) #      .
```

```python
vocab_size = len(t.word_index) + 1
print("=3")
```

```python
sub_text = ['', '', '', '']
encoded = t.texts_to_sequences([sub_text])
print(encoded)
```

```python
one_hot = to_categorical(encoded, num_classes = vocab_size)
print(one_hot)
```

### 7-8 Word2Vec (4)  Word2Vec  OOV  -->  

```python
import nltk
nltk.download('abc')
nltk.download('punkt')
```

```python
from nltk.corpus import abc
corpus = abc.sents()
print("~")
```

```python
print(corpus[:3])
```

```python
print('  :',len(corpus))
```

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences = corpus, vector_size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
print("  !")
```

```python
model_result = model.wv.most_similar("man")
print(model_result)
```

```python
from gensim.models import KeyedVectors

model.wv.save_word2vec_format('~/aiffel/word_embedding/w2v') 
loaded_model = KeyedVectors.load_word2vec_format("~/aiffel/word_embedding/w2v")
print("  load !")
```

```python
model_result = loaded_model.most_similar("man")
print(model_result)
```
